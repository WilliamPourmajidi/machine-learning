Even when a scatter plot looks clear to the human eye, clustering algorithms can still behave differently because they use different ideas:

- **K-Means** looks for compact centroid-based groups.
- **DBSCAN** looks for dense regions and can mark some points as noise.
- **Agglomerative clustering** builds clusters by merging nearby points step by step.

## 2. Create a simple 2D student dataset

We will generate a synthetic dataset that represents **student weekly habits**:

- `StudyHours`
- `SocialHours`

The data will contain a few natural behavior patterns plus a small set of unusual points.

### Suggested interpretation of the groups

You may eventually notice patterns such as:

- students who study a lot and socialize less,
- students who are balanced,
- students who socialize more and study less,
- a few unusual outliers.

Your task is not to force those labels at the beginning.  
Instead, let the clustering results guide your interpretation.

```python
# Create three student lifestyle groups plus a few outliers

n1, n2, n3 = 45, 45, 45

# Group 1: higher study, lower social hours
g1_study  = np.random.normal(loc=22, scale=3.0, size=n1)
g1_social = np.random.normal(loc=6,  scale=1.8, size=n1)

# Group 2: balanced students
g2_study  = np.random.normal(loc=14, scale=2.5, size=n2)
g2_social = np.random.normal(loc=12, scale=2.3, size=n2)

# Group 3: lower study, higher social hours
g3_study  = np.random.normal(loc=7,  scale=2.0, size=n3)
g3_social = np.random.normal(loc=19, scale=2.8, size=n3)

# Outliers / unusual habits
out_study  = np.array([3, 28, 26, 1, 30, 10])
out_social = np.array([2, 24, 3, 26, 1, 28])

study_hours = np.concatenate([g1_study, g2_study, g3_study, out_study])
social_hours = np.concatenate([g1_social, g2_social, g3_social, out_social])

df = pd.DataFrame({
    "StudyHours": np.clip(study_hours, 0, None),
    "SocialHours": np.clip(social_hours, 0, None)
})

df.head()
```
