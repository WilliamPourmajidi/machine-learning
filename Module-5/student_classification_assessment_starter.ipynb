```python
# ============================================================
# STEP 2: Generate the synthetic dataset
# ============================================================

# This dataset simulates student academic behavior and performance.
# Each row represents one student.

# IMPORTANT:
# Do not change the random seed. This ensures everyone works with the same dataset.

np.random.seed(42)

# Number of students
n = 400

# Generate feature columns
study_hours_per_week = np.random.randint(0, 26, n)          # 0 to 25 hours
attendance_rate = np.random.randint(40, 101, n)             # 40% to 100%
assignment_avg = np.random.randint(35, 101, n)              # 35 to 100
quiz_avg = np.random.randint(30, 101, n)                    # 30 to 100
lab_performance = np.random.randint(30, 101, n)             # 30 to 100
participation_score = np.random.randint(0, 11, n)           # 0 to 10
previous_gpa = np.round(np.random.uniform(1.5, 4.0, n), 2)  # 1.5 to 4.0

# Create a weighted score to simulate pass / need support
# The target should depend on a combination of features, not just one variable.
performance_score = (
    0.10 * study_hours_per_week +
    0.20 * attendance_rate +
    0.20 * assignment_avg +
    0.15 * quiz_avg +
    0.15 * lab_performance +
    1.5 * participation_score +
    8 * previous_gpa +
    np.random.normal(0, 8, n)   # noise term to make the problem more realistic
)

# Create binary target
# Students above the median score will be labeled as Pass (1)
# Students below the median score will be labeled as Need Support (0)
final_outcome = (performance_score > np.median(performance_score)).astype(int)

# Create a pandas DataFrame
df = pd.DataFrame({
    "study_hours_per_week": study_hours_per_week,
    "attendance_rate": attendance_rate,
    "assignment_avg": assignment_avg,
    "quiz_avg": quiz_avg,
    "lab_performance": lab_performance,
    "participation_score": participation_score,
    "previous_gpa": previous_gpa,
    "final_outcome": final_outcome
})

# Display the first 5 rows
# TODO:
# Show the first 5 rows of the dataset
```
